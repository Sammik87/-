```python
service = Service(executable_path = 'C:\Python\chromedriver.exe')
options = webdriver.ChromeOptions()

options.add_argument("--disable-blink-features")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument("start_maximized")

driver = webdriver.Chrome(service = service, options = options)

stealth(driver,
#        user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.53 Safari/537.36',
        languages=["en-US", "en"],
        vendor = "Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=False, #True 
        run_on_insecure_origins=False
        )

pages_lnk = []
t1 = [] #список
a = 0

for lnk in tqdm(link_all, 'Сбор данных'):
    try:
        a = a + 1 #счетчик
        t1.clear() #очищаем список значений
        #url = f"https://vl.etagi.com/realty/10106237/"
        driver.get(lnk)
        sleep(3)

    #         b = 1
    #         while b == 1:
    #             try:
    #                 button = driver.find_element(By.XPATH, '/html/body/div[7]/div/div[1]/button') #закрываем выбор языка
    #                 button.click() #нажимаем кнопку крестик
    #             except:
    #                 #pass
    #                 continue
    #             b = b + 1

        try:
            button = driver.find_element(By.XPATH, '//*[@id="about"]/div[3]/div/div[2]/p/span[2]/button') #находим элемент кнопки Еще
            button.click() #нажимаем кнопку Еще
        except:
            #pass
            continue

        try:
            button = driver.find_element(By.XPATH, '//*[@id="characteristics"]/div/button') #находим элемент кнопки Показать
            button.click() #нажимаем кнопку Показать
        except:
            #pass
            continue

        #ХАРАКТЕРИСТИКИ_1
        inf_1 = driver.find_element(By.CLASS_NAME, "iUYKQ").text #.split(",")
        inf_1_s = ['Характеристики', inf_1] #создаем список
        t1.append(inf_1_s)                  #складываем в список

        #АДРЕС
        addres = driver.find_element(By.CLASS_NAME, "Vs6MQ").text #.split(",")
        addres_s = ['Адрес', addres] #создаем список
        t1.append(addres_s)          #складываем в список

        #ЦЕНА
        price = driver.find_element(By.CLASS_NAME, "fNLRl").text #.split(",")
        price_s = ['Цена', price] #создаем список
        t1.append(price_s)        #складываем в список

        #ЦЕНА первоначальная
        try:
            price_2 = driver.find_element(By.CLASS_NAME, "iYh9z").text #.split(",")
            price_2_s = ['Цена_перв', price_2] #создаем список
            t1.append(price_2_s)               #складываем в список
        except:
            price_2_s = ['Цена_перв', 0]
            t1.append(price_2_s)
            #continue

        #ПРОСМОТРЫ
        try:
            see = driver.find_element(By.CLASS_NAME, "rabOe").text #.split(",")
            see_s = ['Просмотры', see] #создаем список
            t1.append(see_s)           #складываем в список
        except:
            see_s = ['Просмотры', 0]
            t1.append(see_s)
            #continue

        #ТЕКСТ
        text = driver.find_element(By.CLASS_NAME, "i5g3k").text #.split(",")
        text_s = ['Текст', text] #создаем список
        t1.append(text_s)        #складываем в список

        #ХАРАКТЕРИСТИКИ_2
        Z = driver.find_elements(By.CLASS_NAME, "Y65Dj")
        inf_2_z = [i.text for i in Z]
        N = driver.find_elements(By.CLASS_NAME, "cfF2L")
        inf_2_n = [j.text for j in N]
        Dictionary = dict(zip(inf_2_z, inf_2_n))
        resultList = list(map(list, Dictionary.items()))
        #t1.append(resultList)
        for res in resultList:
            t1.append(res)

        #ССЫЛКА
        lnk_1 = ['Ссылка', lnk] #создаем список
        t1.append(lnk_1)        #складываем в список

        df = pd.DataFrame(t1)
        df = df.set_index(0)    

        if a == 1:
            df_result = df
        if a > 1:
            df_result = pd.concat([df_result, df], axis = 1)

    except:
        winsound.Beep(2000, 3000) #громкий звук в 3 секунд
        print('Сплю 10 сек')
        pages_lnk.append(lnk)     #страницы, которые не прошли
        sleep(10)                 #спим 10 сек
        #continue

df_result_copy = df_result.copy()
df_result_copy = df_result_copy.transpose()

#Сохраняем в Excel
writer = pd.ExcelWriter('Квартиры_Владивосток_вторич_этажи_2.xlsx')
df_result_copy.to_excel(writer, 'marks')
writer.save()
```
