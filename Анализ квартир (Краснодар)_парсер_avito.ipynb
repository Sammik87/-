```python
service = Service(executable_path = 'C:\Python\chromedriver.exe')
options = webdriver.ChromeOptions()

options.add_argument("--disable-blink-features")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument("start_maximized")

driver = webdriver.Chrome(service = service, options = options)

stealth(driver,
#        user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.53 Safari/537.36',
        languages=["en-US", "en"],
        vendor = "Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=False, #True 
        run_on_insecure_origins=False
        )

a = 0
p = 74 #количество страниц + 1, 48, 74

for page in tqdm(range(1, p, 1), 'Обработанно страниц'):
    a = a + 1 #счетчик
    url = f"https://www.avito.ru/krasnodar/kvartiry/prodam/novostroyka-ASgBAgICAkSSA8YQ5geOUg?cd=1&context=H4sIAAAAAAAA_0q0MrSqLraysFJKK8rPDUhMT1WyLrYyt1JKTixJzMlPV7KuBQQAAP__dhSE3CMAAAA&f=ASgBAQECBESSA8YQ5geOUsibDLLFF8DBDbr9NwFA5AdUluuZAv5R_FH6UfhRAUW4wQ0UeyJmcm9tIjpudWxsLCJ0byI6MX0&p={page}" #вторичка 2
    #чист - https://www.avito.ru/krasnodar/kvartiry/prodam/novostroyka/s-remontom-ASgBAgICA0SSA8YQ5geOUsibDLTFFw?cd=1&context=H4sIAAAAAAAA_0q0MrSqLraysFJKK8rPDUhMT1WyLrYyt1JKTixJzMlPV7KuBQQAAP__dhSE3CMAAAA&f=ASgBAQECBESSA8YQ5geOUsibDLTFF8DBDbr9NwFA5AdUluuZAv5R_FH6UfhRAUW4wQ0UeyJmcm9tIjpudWxsLCJ0byI6MX0&p=4
    #пред чист - https://www.avito.ru/krasnodar/kvartiry/prodam/novostroyka-ASgBAgICAkSSA8YQ5geOUg?cd=1&context=H4sIAAAAAAAA_0q0MrSqLraysFJKK8rPDUhMT1WyLrYyt1JKTixJzMlPV7KuBQQAAP__dhSE3CMAAAA&f=ASgBAQECBESSA8YQ5geOUsibDLLFF8DBDbr9NwFA5AdUluuZAv5R_FH6UfhRAUW4wQ0UeyJmcm9tIjpudWxsLCJ0byI6MX0&p=4
    driver.get(url)
    sleep(5) #спим

    blocks = driver.find_elements(By.CLASS_NAME, 'iva-item-body-KLUuy')
    
    t1 = [] #список
    for block in blocks:
        t1.clear() #очищаем список значений
        
        #Цена
        try:
            price = block.find_element(By.CLASS_NAME, 'price-price-JP7qe').text
            price_ = ['Цена', price]
            t1.append(price_)
        except:
            price_ = ['Цена', 'no']
            t1.append(price_)
        
        #Инфа о квартире
        try:
            info = block.find_element(By.CLASS_NAME, 'iva-item-titleStep-pdebR').text
            info_ = ['Инфо', info]
            t1.append(info_)
        except:
            info_ = ['Инфо', 'no']
            t1.append(info_)
        
        #Адрес
        try:
            address = block.find_element(By.CLASS_NAME, 'geo-root-zPwRk').text
            address_ = ['Адрес', address]
            t1.append(address_)
        except:
            address_ = ['Адрес', 'no']
            t1.append(address_)
        
        #Текст
        try:
            text = block.find_element(By.CLASS_NAME, 'iva-item-descriptionStep-C0ty1').text
            text_ = ['Текст', text]
            t1.append(text_)
        except:
            text_ = ['Текст', 'no']
            t1.append(text_)
            
        #Отделка
        remont = ['Ремонт', 'предчистовая']
        t1.append(remont)
                
        #Ссылка
        try:
            link = block.find_element(By.CLASS_NAME, 'iva-item-titleStep-pdebR').find_element(By.TAG_NAME, 'a').get_attribute("href")
            link_ = ['Ссылка', link]
            t1.append(link_)
        except:
            link_ = ['Ссылка', 'no']
            t1.append(link_)
            
        df = pd.DataFrame(t1)
        df = df.set_index(0)    

        if a == 1:
            df_result = df
        if a > 1:
            df_result = pd.concat([df_result, df], axis=1)

df_result_copy = df_result.copy()
df_result_copy = df_result_copy.transpose()
#Сохраняем в Excel
writer = pd.ExcelWriter('Крас_кв_предчист_avito.xlsx')
df_result_copy.to_excel(writer, 'link')
writer.save()
```
