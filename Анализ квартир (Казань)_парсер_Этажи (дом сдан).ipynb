```python
service = Service(executable_path = 'C:\Python\chromedriver.exe')
options = webdriver.ChromeOptions()

options.add_argument("--disable-blink-features")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument("start_maximized")

driver = webdriver.Chrome(service = service, options = options)

stealth(driver,
#        user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.53 Safari/537.36',
        languages=["en-US", "en"],
        vendor = "Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=False, #True 
        run_on_insecure_origins=False
        )

pages_lnk = []
t1 = [] #список
a = 0

for lnk in tqdm(link_all_2, 'Сбор данных'):
    try:
        a = a + 1 #счетчик
        t1.clear() #очищаем список значений
        #url = f"https://vl.etagi.com/realty/10106237/"
        driver.get(lnk)
        sleep(3)

#         b = 1
#         while b == 1:
#             try:
#                 button = driver.find_element(By.XPATH, '/html/body/div[7]/div/div[1]/button') #закрываем выбор языка
#                 button.click() #нажимаем кнопку крестик
#             except:
#                 #pass
#                 continue
#             b = b + 1

        #ХАРАКТЕРИСТИКИ_1
        inf_1 = driver.find_element(By.CLASS_NAME, "JGoRb").text #.split(",")
        inf_1_s = ['Характеристики', inf_1] #создаем список
        t1.append(inf_1_s)                  #складываем в список

        #ЦЕНА
        price = driver.find_element(By.CLASS_NAME, "eypL8").text #.split(",")
        price_s = ['Цена', price] #создаем список
        t1.append(price_s)        #складываем в список

        #ХАРАКТЕРИСТИКИ_2
        Z = driver.find_elements(By.CLASS_NAME, "RzVYb")
        inf_2_z = [i.text for i in Z]
        N = driver.find_elements(By.CLASS_NAME, "pXUis")
        inf_2_n = [j.text for j in N]
        Dictionary = dict(zip(inf_2_z, inf_2_n))
        resultList = list(map(list, Dictionary.items()))
        #t1.append(resultList)
        for res in resultList:
            t1.append(res)

        #ССЫЛКА
        lnk_1 = ['Ссылка', lnk] #создаем список
        t1.append(lnk_1)        #складываем в список

        df = pd.DataFrame(t1)
        df = df.set_index(0)    

        if a == 1:
            df_result = df
        if a > 1:
            df_result = pd.concat([df_result, df], axis = 1)

    except:
        winsound.Beep(2000, 2000) #громкий звук в 2 секунд
        print('Сплю 10 сек')
        pages_lnk.append(lnk)     #страницы, которые не прошли
        sleep(10)                 #спим 10 сек
        #continue

df_result_copy = df_result.copy()
df_result_copy = df_result_copy.transpose()

#Сохраняем в Excel
writer = pd.ExcelWriter('Квартиры_Казань_этажи_нов.xlsx')
df_result_copy.to_excel(writer, 'marks')
writer.save()
```
